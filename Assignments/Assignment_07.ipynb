```python
def assign_quartile(gdp):
    if gdp <= 1862.89:
        quartile = "1st Quartile (<=25%)"
    elif gdp > 1862.89 and gdp <= 6048.30:
        quartile = "2nd Quartile (25-50%)"
    elif gdp > 6048.30 and gdp <= 22137.31:
        quartile = "3rd Quartile (50-75%)"
    else:
        quartile = "4th Quartile (>75%)"
    return quartile


hungary = assign_quartile(14294.25)
hungary
```

## Homework Question 3 (bonus, will cover next week)

step 1 - create a column "Life Expectancy Quartile" similar to the GDP per Capita Quartile above.
Step 2 - Create a column "health and wealth status" based on the following definition
- rich and healthy (1st GDP per capita quartile and 1st life expentancy quartile)
- poor and unhealthy ((4th GDP per capita quartile and 4th life expentancy quartile)
- rich and unhealthy (1st GDP per capita quartile and 4th life expentancy quartile)
- poor and healthy ((4th GDP per capita quartile and 1st life expentancy quartile)
- Other

```python
#Defining a function for life expectancy quartile
def life_expectancy_quartile(life_exp):
    if life_exp <= 60:
        return "1st Quartile (<=25%)"
    elif 60 < life_exp <= 70:
        return "2nd Quartile (25-50%)"
    elif 70 < life_exp <= 75:
        return "3rd Quartile (50-75%)"
    else:
        return "4th Quartile (>75%)"
```

```python
# Applying the function to create the Life Expectancy Quartile column
df_2020["Life Expectancy Quartile"] = df_2020["Life Expectancy"].apply(life_expectancy_quartile)
```

```python
# Creating the Health and Wealth Status  column
def health_wealth_status(row):
    if row["GDP Quartile"] == "1st Quartile (<=25%)" and row["Life Expectancy Quartile"] == "1st Quartile (<=25%)":
        return "Rich and Healthy"
    elif row["GDP Quartile"] == "4th Quartile (>75%)" and row["Life Expectancy Quartile"] == "4th Quartile (>75%)":
        return "Poor and Unhealthy"
    elif row["GDP Quartile"] == "1st Quartile (<=25%)" and row["Life Expectancy Quartile"] == "4th Quartile (>75%)":
        return "Rich and Unhealthy"
    elif row["GDP Quartile"] == "4th Quartile (>75%)" and row["Life Expectancy Quartile"] == "1st Quartile (<=25%)":
        return "Poor and Healthy"
    else:
        return "Other"
```

```python
# Apply the function to create the "Health and Wealth Status" column
df_2020["Health and Wealth Status"] = df_2020.apply(health_wealth_status, axis=1)
df_2020.head(10)
```
