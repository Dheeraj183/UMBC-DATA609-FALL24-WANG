```python
import pandas as pd
import plotly.express as px

pd.set_option("display.max_rows", None)

df = pd.read_csv("https://raw.githubusercontent.com/wcj365/python-stats-dataviz/refs/heads/master/fall2024/data/World_Development_Indicators_(WDI).csv")

df.shape
```

```python
df.sample(5)
```

```python
df.isna().sum()
```

```python
df[["Year", "Country","GDP per capita (current US$)", "Life expectancy at birth, total (years)"]].sample(5)
```

```python
df2 = df[df["Year"] != 2022]
df2.shape
```

```python
df2.isna().sum()
```

```python
df3 = df2.dropna()
df3.isna().sum()
```

```python
df3.shape
```

## Homework Question 1:

to check if this dataset has duplicate rows. If it has, only keep one from the duplicated rows and drop the rest.

```python
# Checking duplicate rows
duplicate_count = df3.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# If duplicates exist, drop them; otherwise, keep the original dataframe
if duplicate_count > 0:
    df3_cleaned = df3.drop_duplicates()
    print(f"Shape after removing duplicates: {df3_cleaned.shape}")
else:
    df3_cleaned = df3  # No duplicates, so just assigning the original dataframe
    print("No duplicate rows found.")

print(f"Number of duplicates after cleaning: {df3_cleaned.duplicated().sum()}")

```

## HW Question 2

Find out if there is a relationship between wealth and health

Pick any year worth of data and use plotly express to generate a scatter plot with GDP as x and life expectancy as y.

```python
import plotly.express as px

# taking 2020
df_2020 = df3[df3["Year"] == 2020]

# Creating a scatter plot to show the relationship between GDP per capita and life expectancy
fig = px.scatter(
    df_2020,
    x="GDP per capita (current US$)",
    y="Life expectancy at birth, total (years)",
    title="Relationship between GDP per capita and Life Expectancy (2020)",
    labels={
        "GDP per capita (current US$)": "GDP per capita (US$)",
        "Life expectancy at birth, total (years)": "Life Expectancy (Years)"
    },
    trendline="ols"  # Adding a trendline
)

fig.show()
```

## Homework Question 3

Select a year of data. calculate total population of each region and make a pie chart to display the distribution of total population of all regions.

```python
import plotly.express as px

# taking 2019
df_2019 = df3[df3["Year"] == 2019]

# Group by region and calculating total population for each region
df_population_by_region = df_2019.groupby("Region")["Population, total"].sum().reset_index()


fig = px.pie(
    df_population_by_region,
    values="Population, total",
    names="Region",
    title="Population Distribution by Region (2019)",
    color="Region",  # Color by region
    color_discrete_sequence=px.colors.qualitative.Pastel  #adding pastel colors
)


fig.update_traces(textinfo='percent+label')

fig.show()
```

## Home work question 4

1. remove the color legend.
2. Order the above bar chart based on the values of the average GDP per capita.

```python
import plotly.express as px

# Sort the data by GDP per capita in descending order
df_group_sorted = df_group.sort_values(by="GDP per capita (current US$)", ascending=False)


fig = px.bar(
    df_group_sorted,
    x="Region",
    y="GDP per capita (current US$)",
    title="Average GDP per capita by Region (Ordered)",
    labels={"GDP per capita (current US$)": "Average GDP per capita (US$)"},
    color_discrete_sequence=["#636EFA"]  # Uniform color for all bars
)

# Removing the legend
fig.update_layout(showlegend=False)
fig.update_traces(marker_line_width=0, width=0.6)
fig.show()
```
