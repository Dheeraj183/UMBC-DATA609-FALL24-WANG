## 10/9/20204 Homework Question 1:

for the Quad time series comparison chart above, try to add country name to the end of the line and then remove the color legend.

```python
countries = ["United States", "China", "India", "Japan"]
df_filtered = df[df["Country"].isin(countries)]

# Creating the line chart
fig = px.line(df_filtered,
              x="Year",
              y="GDP per capita (current US$)",
              color="Country",
              title="GDP per Capita Over Time for Quad Countries")

# Removing the legend
fig.update_layout(showlegend=False)

# Adding country names at the end of each line
for country in countries:

    country_data = df_filtered[df_filtered["Country"] == country]
    last_year = country_data["Year"].max()
    last_value = country_data[country_data["Year"] == last_year]["GDP per capita (current US$)"].values[0]


    fig.add_annotation(x=last_year,
                       y=last_value,
                       text=country,
                       showarrow=False,
                       font=dict(size=12, color=fig.data[countries.index(country)].line.color))


fig.show()
```

## Question 3 : Bonus

```python
# Group by Year and Region, calculate average GDP per capita, average life expectancy, and total population
df_region_agg = df.groupby(["Year", "Region"]).agg(
    {
        "GDP per capita (current US$)": "mean",
        "Life expectancy at birth, total (years)": "mean",
        "Population, total": "sum"
    }
).reset_index()

df_region_agg.sample(5)
```

```python
# Create a line chart showing the trend of Life Expectancy over time for all regions
fig = px.line(
    df_region_agg,
    x="Year",
    y="Life expectancy at birth, total (years)",
    color="Region",
    title="Life Expectancy Trend for All Regions",
    labels={
        "Year": "Year",
        "Life expectancy at birth, total (years)": "Life Expectancy (Years)"
    }
)

fig.show()


```

```python
# Pick a year from the data and create a scatter plot to compare their GDP per capita and life expectancy.
year_of_interest = 2019
df_region_agg_year = df_region_agg[df_region_agg["Year"] == year_of_interest]

fig = px.scatter(
    df_region_agg_year,
    x="GDP per capita (current US$)",
    y="Life expectancy at birth, total (years)",
    color="Region",
    size="Population, total",
    title=f"GDP per Capita vs. Life Expectancy for All Regions in {year_of_interest}",
    labels={
        "GDP per capita (current US$)": "GDP per Capita (Current US$)",
        "Life expectancy at birth, total (years)": "Life Expectancy (Years)"
    }
)

fig.show()
```

## 10/09/2024 Homework Question 2

Round the Average GDP for each region to 2 decimal points.
Sort the bar according to GDP either in ascending order. Currently, it is ordered alphabetically.

```python
avg_gdp_per_region = df.groupby("Region")["GDP per capita (current US$)"].mean().round(2).reset_index()

# Sort the data by average GDP in ascending order
avg_gdp_per_region = avg_gdp_per_region.sort_values("GDP per capita (current US$)", ascending=True)

# Creating a bar chart
fig = px.bar(avg_gdp_per_region,
             x="Region",
             y="GDP per capita (current US$)",
             title="Average GDP per Capita by Region (Sorted)")


fig.update_layout(xaxis_title="Region", yaxis_title="Average GDP per Capita (US$)", showlegend=False)

fig.show()
```

## 10/09/2024 Homework question 3

Group the dataset by Region and Income Group, calculate the average, min, and max of GDP per Capita & life Expentancy, total, average, min, and max of population, then save the result to Excel.


```python
grouped_stats = df.groupby(["Region", "Income Group"]).agg(
    avg_gdp_per_capita=("GDP per capita (current US$)", "mean"),
    min_gdp_per_capita=("GDP per capita (current US$)", "min"),
    max_gdp_per_capita=("GDP per capita (current US$)", "max"),
    avg_life_expectancy=("Life expectancy at birth, total (years)", "mean"),
    min_life_expectancy=("Life expectancy at birth, total (years)", "min"),
    max_life_expectancy=("Life expectancy at birth, total (years)", "max"),
    avg_population=("Population, total", "mean"),
    min_population=("Population, total", "min"),
    max_population=("Population, total", "max")
).reset_index()

# Saving the result to an Excel file
grouped_stats.to_excel("Grouped_Stats_by_Region_and_Income_Group.xlsx", index=False)

print("The grouped statistics have been saved to 'Grouped_Stats_by_Region_and_Income_Group.xlsx'.")
```
