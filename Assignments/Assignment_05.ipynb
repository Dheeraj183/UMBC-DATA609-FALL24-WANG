Pick a few countries of your interest (for example, G7, BRICS, Nordic countries).
- 1 Create a line chart showing the trend of Life Expectancy over time for comparing those countries.
- 2 Pick a year from the data of the above few countries and create a scatter plot to compare their GDP per capita and life expectancy. The size of the dots represent the population. This type of scatter plot is called bubble chart.
- 3 Bonus - compute the average GDP per capita, the average life expectancy, and the total population for each year of all the seven regions (aggregation) and repeat 1 and 2 using the aggregated data for all regions

**1 Create a line chart showing the trend of Life Expectancy over time for comparing those countries.**

```python
# Filter data for selected G7 countries
g7_countries = ["USA", "CAN", "GBR", "DEU", "FRA", "ITA", "JPN"]

df_g7 = df[df["Country Code"].isin(g7_countries)]

# Create a line chart to show the trend of Life Expectancy over time for the selected countries
fig = px.line(
    df_g7,
    x="Year",
    y="Life expectancy at birth, total (years)",
    color="Country",
    title="Life Expectancy Trend Over Time for G7 Countries",
    labels={
        "Life expectancy at birth, total (years)": "Life Expectancy (Years)",
        "Year": "Year"
    }
)

fig.show()
```

**2 Pick a year from the data of the above few countries and create a scatter plot to compare their GDP per capita and life expectancy. The size of the dots represent the population. This type of scatter plot is called bubble chart.**

```python
# Select data for the year 2019 and the selected G7 countries
year_of_interest = 2019
df_g7_2019 = df[(df["Year"] == year_of_interest) & (df["Country Code"].isin(g7_countries))]

# Drop rows with missing data for GDP or Life Expectancy
df_g7_2019 = df_g7_2019.dropna(subset=["GDP per capita (current US$)", "Life expectancy at birth, total (years)"])

# Create a bubble chart comparing GDP per capita and Life Expectancy
fig = px.scatter(
    df_g7_2019,
    x="GDP per capita (current US$)",
    y="Life expectancy at birth, total (years)",
    size="Population, total",
    color="Country",
    title=f"GDP per Capita vs Life Expectancy for G7 Countries in {year_of_interest}",
    labels={
        "GDP per capita (current US$)": "GDP per Capita (Current US$)",
        "Life expectancy at birth, total (years)": "Life Expectancy (Years)"
    }
)

fig.show()
```

**3 Bonus - compute the average GDP per capita, the average life expectancy, and the total population for each year of all the seven regions (aggregation) and repeat 1 and 2 using the aggregated data for all regions**

```python
# Group data by Year and Region, and compute the average GDP per capita, life expectancy, and total population
df_region_agg = df.groupby(["Year", "Region"]).agg(
    avg_gdp=("GDP per capita (current US$)", "mean"),
    avg_life_expectancy=("Life expectancy at birth, total (years)", "mean"),
    total_population=("Population, total", "sum")
).reset_index()

# Line chart showing the trend of Life Expectancy over time by region
fig = px.line(
    df_region_agg,
    x="Year",
    y="avg_life_expectancy",
    color="Region",
    title="Life Expectancy Trend Over Time by Region",
    labels={
        "avg_life_expectancy": "Average Life Expectancy (Years)",
        "Year": "Year"
    }
)

fig.show()

```

```python
# Select a specific year for the scatter plot (e.g., 2019)
df_region_2019 = df_region_agg[df_region_agg["Year"] == year_of_interest]

# Create a bubble chart comparing GDP per capita and Life Expectancy for regions
fig = px.scatter(
    df_region_2019,
    x="avg_gdp",
    y="avg_life_expectancy",
    size="total_population",
    color="Region",
    title=f"Average GDP per Capita vs Life Expectancy by Region in {year_of_interest}",
    labels={
        "avg_gdp": "Average GDP per Capita (Current US$)",
        "avg_life_expectancy": "Average Life Expectancy (Years)"
    }
)

fig.show()
```
